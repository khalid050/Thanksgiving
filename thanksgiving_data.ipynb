```python
#number of people that celebrate thanksgiving
celebrate_column = data.iloc[:,1]
celebrate_column.value_counts()
```

```python
#altering data to have only rows where respondants answered 'yes' to celebrating thanksgiving 
data = data[data.iloc[:,1] == 'Yes']
data.head()
```

```python
#Main dish people eat during thanksgiving 
main_dish_col = data['What is typically the main dish at your Thanksgiving dinner?']
main_dish_col.value_counts()
```

```python
# A function that returns the boolean value of weather or no the respondant ate a particular pie. 
def null_values(column):
    null = pd.isnull(data[column])
    return null
```

```python
apple_null = null_values('Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple')
apple_null.value_counts()
    
    
```

```python
pumpkin_null = null_values('Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin')
pumpkin_null.value_counts()
```

```python
pecan_null = null_values('Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan')
pecan_null.value_counts()
```

```python
# Number of people that eat all three pies during thanksgiving
no_pies = apple_null & pumpkin_null & pecan_null
no_pies.value_counts()
```

```python
#A function that returns the first age from the range in the 'Age' column and converts the age into an integer
def extract_age(age):
    if pd.isnull(age):
        return None
    age = age.split(" ")[0]
    age = age.replace("+","")
    return int(age)
```

```python
data['int_age'] = data['Age'].apply(extract_age)
data['int_age'].describe()
```

# Findings

The age group is evenly distributed despite using the first number in the age string

```python
def extract_income(income):
    if pd.isnull(income):
        return None
    income = income.split(" ")[0]
    if income == 'Prefer':
        return None
    income = income.replace('$' , '')
    income = income.replace(',' , '')
    return int(income)
```

```python
data['int_income'] = data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(extract_income)
data['int_income'].describe()
```

```python
data[data['int_income'] <150000]['How far will you travel for Thanksgiving?'].value_counts()
```

```python
data[data['int_income']> 150000]['How far will you travel for Thanksgiving?'].value_counts()
```

# Findings

We can see that those with lower income are more inclined to travel for thanksgiving than those with higher income. This could because younger people with lower income may go home for thanksgiving, given their income

```python
dessert_columns = data.loc[:,'Which of these desserts do you typically have at Thanksgiving dinner? Please select all that apply.   - Apple cobbler' 
                           : 'Which of these desserts do you typically have at Thanksgiving dinner? Please select all that apply.   - Peach cobbler']
dessert_columns.head()

```

```python
dessert_count = {}
for column in dessert_columns.columns:
    dessert_name = column.split(' ')[-1]
    dessert_count[dessert_name] = pd.notnull(dessert_columns[column]).value_counts()[True]

#renaming the names of dessert the function could not capture
dessert_count['Carrot cake'] = dessert_count.pop('cake')
dessert_count['Icecream'] = dessert_count.pop('cream')

dessert_count

```

# Findings

Icecream is the most popular dessert choice for Thanksgiving
